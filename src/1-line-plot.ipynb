### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b')

plt.title('Square of Numbers from 0 to 10')
plt.xlabel('X values')
plt.ylabel('Y = X^2')
plt.grid(True)

plt.show()
```
