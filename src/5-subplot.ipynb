### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data for all plots
line_x = [0, 1, 2, 3, 4]
line_y = [0, 1, 4, 9, 16]

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]


fig, axs = plt.subplots(2, 2, figsize=(10, 10))


axs[0, 0].plot(line_x, line_y, marker='o')
axs[0, 0].set_title('Line Chart')


axs[0, 1].bar(bar_categories, bar_values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')


axs[1, 0].scatter(scatter_x, scatter_y, color='purple')
axs[1, 0].set_title('Scatter Plot')

axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio for circular pie

plt.tight_layout()
plt.show()
```
